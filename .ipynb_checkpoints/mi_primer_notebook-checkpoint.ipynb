#### Segundo código - demo_seg_orden.py

A continuación, se prueba el código provisto por la cátedra en distintos pasos.

Primero, se importan las librerías:

```python
# Importo las librerías
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
```

Luego, se genera la función transferencia a partir de los parámetros $\omega_0$ y $Q$:

```python
w0 = 1
qq = np.sqrt(2)/2

my_tf = TransferFunction( [w0**2], [1, w0/qq, w0**2] )
```

Finalmente, se generan los gráficos de módulo, fase, delay de grupo y diagrama de polos y ceros para esta función transferencia:

```python
bodePlot(my_tf, fig_id=1, filter_description = 'Q={:3.3f}'.format(qq) )

pzmap(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(qq)) #S plane pole/zero plot

GroupDelay(my_tf, fig_id=3, filter_description = 'Q={:3.3f}'.format(qq))
```

#### Sistema amortiguado crítico

En el ejemplo provisto, el sistema encontrado está en régimen subamortiguado. Esto se puede observar, por ejemplo, en el diagrama de polos y ceros, donde se encuentran dos raíces complejas con parte real negativa.

Si se modifica el valor del factor de calidad $Q$, de forma que $Q = 1/2$, tendremos un **sistema amortiguado crítico**, lo cual también se podrá observar mediante el diagrama de polos y ceros, donde encontraremos una raíz real pura doble.

Para esto, se modifica únicamente la variable *qq*, haciendo que valga 0,5.

```python
qq = 1/2

my_tf = TransferFunction( [w0**2], [1, w0/qq, w0**2] )

bodePlot(my_tf, fig_id=1, filter_description = 'Q={:3.3f}'.format(qq) )

pzmap(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(qq)) #S plane pole/zero plot

GroupDelay(my_tf, fig_id=3, filter_description = 'Q={:3.3f}'.format(qq))
```

#### Sistema sobreamortiguado

De igual manera, si se modifica el factor de calidad de forma que $Q < 1/2$, se obtendrá un **sistema sobreamortiguado**, lo cual se reflejará en dos raíces reales puras negativas.

Para esto, se le da un valor a la variable *qq* de 0,25.


```python
w0 = 1
qq = 1/4

my_tf = TransferFunction( [w0**2], [1, w0/qq, w0**2] )

bodePlot(my_tf, fig_id=1, filter_description = 'Q={:3.3f}'.format(qq) )

pzmap(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(qq)) #S plane pole/zero plot

GroupDelay(my_tf, fig_id=3, filter_description = 'Q={:3.3f}'.format(qq))
```
